```python
##A 공장의 2019-01-01 부터 10일간 제품 생산량을 Series에 저장 - 
##단, 생산량의 평균은 50 이고 표준편차는 5인 정규분포에서 생산량을 랜덤하게 결정

##B 공장의 2019-01-01 부터 10일간 제품 생산량을 Series에 저장 - 
##단, 생산량의 평균은 70 이고 표준편차는 8인 정규분포에서 생산량을 랜덤하게 결정


##2019-01-01부터 10일간 모든 공장(A,B)의 생산량의 합을 구하세요~
#날짜별로 출력하세요!



import numpy as np  #필요할것같은 패키지/모듈을 쓰세요
import pandas as pd
from datetime import date,timedelta
from dateutil.parser import parse


start_day = parse("2019-01-01")
factory_a = pd.Series([int(x) for x in np.random.normal(50,5,(10,))],
                      index=[start_day + timedelta(days=x) for x in range (10)])
factory_b = pd.Series([int(x) for x in np.random.normal(70,8,(10,))],
                      index=[start_day + timedelta(days=x) for x in range (10)])

display(factory_a)
display(factory_b)

display(factory_a + factory_b)
```

```python
##평균(mean) : 수학적 확률의 기댓값
#              어떤 사건을 무한히 반복했을 때 얻을 수 있는 평균으로서 기대할 수 있는 값



#편차(deviation) : 확률변수 x와 평균과의 차이
#                : 데이터의 흩어진 정도를 수치화 하기에 적합하지 않아요
#                : 편차의 합은 0이기 때문에.

#분산(Variance) : 데이터의 흩어진 정도를 알기 위해서
#                 사용되는 편차의 제곱의 평균
#                 제곱했기 때문에 단위 표현이 적절치 않아짐..애매해져

#표준편차(Standard deviation) : 분산의 제곱근

# 공분산(covariance) : 두개의 확률변수에 대한 관계를 보여주는 값
# 두개의 확률변수에 대한 공분산값이 양수
#  => 하나의 확률변수가 증가할때 다른 확률변수도 증가하는 경향이 있다.
# 두개의 확률변수에 대한 공분산 값이 음수
# 공분산 값이 0 : 두개의 확률변수는 아무런 관련이 없다(두 데이터는 독립)

# 상관관계(corelation) :두 대상이 서로 연관성이 있다고 추측되는 관계
# 성적 vs 자존감
# 온라인 게임 시간 vs 폭력성


#상관계수(corelation coefficient ) :
#                -1 과 1 사이의 실수
# 0에 가까울수록 연관성이 없다고 판단
# 절대값이 1에 가까울수록 밀접한 연관이 있다고 판단


##상관관계는 인과관계를 설명할 수 없어요!
```

```python
import numpy as np
import pandas as pd

data1 = {"학번":[1,2,3,4],
        "이름":["홍길동","김길동","박길동","이길동"],
        "학년":[2,4,1,3]}
data2 = {"학번" : [1,2,3,4],
         "학과" : ["컴퓨터","경영","철학","기계"],
         "학점" : [3.4, 1.9, 4.5, 2.7]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
#display(df1)
#display(df2)

df=pd.merge(df1,df2, on="학번" ,how="inner")  ##pandas가 제공하는 함수로 inner join 할거에요 (학번을 기준으로)
#display(df)

df4=pd.merge(df1,df2, on="학번",how="outer")##아우터 조인으로 할거에요
display(df4)

df5=pd.merge(df1, df2, on="학번", how="left")
display(df5)
```

```python
import numpy as np
import pandas as pd
data1 = {"학번":[1,2,3,4],
        "이름":["홍길동","김길동","박길동","이길동"],
        "학년":[2,4,1,3]}
data2 = {"학생학번" : [1,2,3,5],
         "학과" : ["컴퓨터","경영","철학","기계"],
         "학점" : [3.4, 1.9, 4.5, 2.7]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
#display(df1)
#display(df2)

##머지할 컬럼의 두개의 데이터는 같은데 column명이 다를때(의미는 같지만)
pd.merge(df1, df2, left_on="학번", right_on="학생학번", how="outer")
#pd.merge(df1, df2, left_on="학번", right_on="학생학번", how="inner")
```

```python
import numpy as np
import pandas as pd
data1 = {"학번":[1,2,3,4],
        "이름":["홍길동","김길동","박길동","이길동"],
        "학년":[2,4,1,3]}
data2 = {
         "학과" : ["컴퓨터","경영","철학","기계"],
         "학점" : [3.4, 1.9, 4.5, 2.7]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2, index=[1,2,4,5])

## key로 사용하는 학번자체가 하나는 key로 있고 하나는 index로 있을 때..
pd.merge(df1,df2,left_on="학번",right_index=True, how="outer") #왼쪽은 학번 컬럼 잡고/// 오른쪽은 인덱스를 기반으로 잡자!

```
