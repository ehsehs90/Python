```python
import numpy as np
import pandas as pd

#2. 각 사용자별 평균 평점을 출력하세요 (for문 / boolean indexing)
df = pd.read_csv("./data/movielens/ratings.csv")
display(df.head())
df2 = pd.read_csv("./data/movielens/movies.csv")
display(df2.head())

#빈 데이터프레임 하나 만들기
result_df=pd.DataFrame(columns=['사용자별 평균평점'])
display(result_df)

#사용자 Id부터 뽑아내 보아요
display(df.head())
#unique를 이용해 중복된걸 제거하고 뽑아낼 거에요
user_ids=df["#userId"].unique()
user_ids.shape     ##610명이구나!

#for문돌려
##610개의 user_id를 하나씩 들고올거에요
for user_id in user_ids:
# #각 유저별로 ratings를 추출 --boolean indexing 사용해가지고
    user_mean = df.loc[df["#userId"] == user_id,"rating"].mean()
    result_df.loc[user_id] = user_mean
    
display(result_df.head())
    

```

```python
#3. 각 영화별 평균 평점을 다음과 같은 형식으로 출력하세요!

#index(영화ID)  영화제목 영화평점
#0                1@#!     3.41
#1                @!#!@    5.43

import numpy as np
import pandas as pd

result_df = pd.DataFrame(columns=["영화제목", "각 영화별 평균평점"])
display(result_df)
# 사용자 id부터 뽑아내 보아요!
df = pd.read_csv("./data/movielens/ratings.csv")
df2 = pd.read_csv("./data/movielens/movies.csv")
df3=pd.merge(df,df2,on="movieId", how="right")

# display(df.head())
movie_ids = df2["movieId"].unique()
# movie_ids.shape
# display(movie_ids)

# df3=df2["title"]
# df4=df.groupby("movieId").mean()["rating"]
# display(pd.merge(df3,df4,left_index=True, right_index=True))

for movie_id in movie_ids:
    movie_mean=df3.loc[df3["movieId"] == movie_id ,"rating"].mean()
    movie_title=df3.loc[df3["movieId"] == movie_id ,"title"].unique()
    result_df.loc[movie_id] = [movie_title, movie_mean]
#display(result_df)
display(result_df.head())
```
