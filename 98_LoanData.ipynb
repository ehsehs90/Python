```python
import warnings
warnings.filterwarnings(action="ignore")

## 1. Data Loading
df = pd.read_csv("./data/loandata/LoanStats3d.csv",
                sep=",",
                skiprows=1)
##csv 불러들이는 셀
```

```python
##총 데이터의 수 확인 (몇 건의 대출정보가 있는지 확인)
display(df.head())
display(df.shape)
display(df.columns)
```

```python
##필요한 column만 추출해서 새로운 DataFrame을 생성
##loan_amnt : 대출금액
##loan_status : 대출상태
## grade : 대출등급
## int_rate : 이자율
## term : 대출기간
df2=df[["loan_amnt","loan_status","grade","int_rate","term"]]   ##fancy indexing//column명 언급
df2.head()

```

```python
##각 컬럼의 값 확인
display(df2["loan_status"].unique())
display(df2["grade"].unique())
display(df2["term"].unique())

##nan 있으면 결측값처리 가야함 (날리든가/다른 값으로 치환하던가)
display(df2.shape)
df2.dropna(how="any", inplace=True)   ##inplace=True:원본을 바꿔 ##how="any"nan있으면 그 줄(row) 날려
display(df2.shape)

##요기까지 하면 분석하기전 준비 완료~
```

```python
##대출기간별 대출총액 구하기


display(df2["term"].unique())




#unique를 이용해 중복된걸 제거하고 뽑아낼 거에요
terms = df2["term"].unique()
term_result = {}  ##빈 딕셔너리 만들기 36를키값/ value값

for t in terms:

    term_sum = df2.loc[df2["term"] == t,"loan_amnt"].sum()
    term_result[t] = term_sum

    #딕셔너리에 결과 넣어가지고 결과값 추출하기.
    
    
for key in term_result:
    print("{}".format(key), end=":")                     ##key 값
    print("{}".format("{:,}".format(term_result[key])))  ##value 값
    
```

```python
##대출 중 불량한 상태의 대출에 대해서
##불량한 상태의 대출(Charged Off, Late(31-120 days)
##(In Grace Period, Late(16-30 days))
##불량한 상태의  총 대출 건수?
##불량한 상태의 대출에 대해 각 Grade의 대출 건수?
##개별적으로 구해보아요


display(df2["loan_status"].unique())



statusbox = df2["loan_status"].unique()
status_result = {}  ##빈 딕셔너리 만들기 키값/ value값
display(statusbox)


for t in statusbox:

    statusbox = df2.loc[df2["loan_status"] == t,"grade"]
    status_result[t] = statusbox

    #딕셔너리에 결과 넣어가지고 결과값 추출하기.
    
    
for key in status_result:
    print("{}".format(key), end=":")
    print("{}".format("{:}".format(status_result[key])))
    
```

```python
##대출 중 불량한 상태의 대출에 대해서
##불량한 상태의 대출(Charged Off, Late(31-120 days)
##(In Grace Period, Late(16-30 days))
##불량한 상태의  총 대출 건수?
##불량한 상태의 대출에 대해 각 Grade의 대출 건수?
##개별적으로 구해보아요

print(df2["loan_status"].unique())
#위에꺼 때리면 이렇게 나옴. 나는 Fully Paid 빼고 다 bad 한 상태다 라고 묶어내고 싶어.
#['Fully Paid' 'Charged Off' 'Current' 'Late (31-120 days)'
# 'In Grace Period' 'Late (16-30 days)']

bad_loan_status=df2["loan_status"].unique()[[1,2,3,4,5]]   ##0번 인덱스 빼고 내가 필요한 인덱스 직접 써서 묶어. 불량한 상태라고.
print(bad_loan_status)

#불량대출에 대한 mask생성(True/False)
bad_mask=df2["loan_status"].isin(bad_loan_status)   ## df2 loan_status 컬럼 내용 중  ban_loan_status에 있는거 mask함

print(bad_mask.sum())

df3=df2.loc[bad_mask,"grade"] #key/ value값
df3.value_counts().sort_index()
```
